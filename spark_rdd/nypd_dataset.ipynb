```python
from datetime import datetime
import csv
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO
from collections import namedtuple
```

```python
#returns field names as list of strings
header.split(",") 
```

```python
header.replace(" ","_").replace("/" , "_") #returns string separated by commas (can verify using type())
```

```python
#returns list of field names
field_names = header.replace(" ","_").replace("/" , "_").split(",") 
field_names
```

```python
#Create Crime class with every index in tuple with field names
Crime = namedtuple('Crime', field_names, verbose = True)
Crime
```

```python
def parse(row1):
    reader = csv.reader(StringIO(row1))
    row1 = next(reader)
    return Crime(*row1)
#returns crime object with every row parsed and combined with field names-value pair   
```

```python
crimes = rows.map(parse)
```
